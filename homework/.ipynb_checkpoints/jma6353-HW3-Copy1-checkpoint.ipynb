```python
import numpy as np
import matplotlib.pyplot as plt
import math

def sinTaylor(x, N):
    """
    Approximates sin(x) using the Taylor Series about 0 with N terms.
    Works when x is a NumPy array.
    """
    total = np.zeros_like(x, dtype=float)
    for n in range(N):
        total += ((-1)**n) * (x**(2*n + 1)) / math.factorial(2*n + 1)
    return total

x = np.arange(0, 10.1, 0.1)

sinExact = np.sin(x)

errors = {}

for numTerms in [1, 2, 3, 4]:
    approx = sinTaylor(x, numTerms)
    errors[numTerms] = np.abs(approx - sinExact)

plt.figure(figsize=(10, 6))
for N in [1, 2, 3, 4]:
    plt.plot(x, errors[N], label=f"{N} term(s)")

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Truncation Error for sin(x)")
plt.legend()
plt.grid(True)
plt.show()

# AI USAGE DISCLOSURE:
# I used AI mainly to double-check my Python syntax and make sure my Taylor series
# implementation was written correctly. I also used it to confirm that the math for
# the Taylor series expansion and truncation matched what I derived.
#
# AI Tool Used: ChatGPT
# AI Prompt Used:
# "Write Python code to approximate sin(x) using a truncated Taylor series with a variable number
# of terms (1â€“4), evaluate for x=0..10 step 0.1, compare to numpy.sin, compute absolute errors,
# plot the error curves, and briefly describe how the error behaves as terms increase."

# The error does not seem to decrease as x increases, it seems to be very accurate when x is near 0.
# It seems like as you keep increasing x, the truncation error seems to increase.
# The 1-4 term approximation become the higher powers of x seem to overpower the neglected remainder.
```
