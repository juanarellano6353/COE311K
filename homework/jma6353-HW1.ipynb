```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(n_samples: int) -> float:
    """
    Estimate pi using a Monte Carlo method.
    Random points are sampled uniformly in the square [-1, 1] x [-1, 1].
    """
    x = np.random.uniform(-1.0, 1.0, n_samples)
    y = np.random.uniform(-1.0, 1.0, n_samples)

    inside_circle = (x**2 + y**2) <= 1.0
    pi_estimate = 4.0 * np.sum(inside_circle) / n_samples

    return pi_estimate


# True value of pi
true_pi = np.pi

# Sample sizes from 10 to 1000 in steps of 20
sample_sizes = range(10, 1001, 20)

errors = []

for n in sample_sizes:
    pi_estimate = monte_carlo_pi(n)
    error = abs(pi_estimate - true_pi)
    errors.append(error)


# ---- Plot AFTER loop finishes ----
plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, errors, marker='o')

plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error |π_est − π_true|")
plt.title("Monte Carlo Pi Approximation Error")
plt.grid(True)

plt.show()
```
