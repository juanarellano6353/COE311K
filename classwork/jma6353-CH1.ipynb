```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 400)
plt.subplot(2, 2, 1)
plt.plot(x,np.sin(x), 'r')
plt.title("sin(x)")

plt.subplot(2, 2, 2)
plt.plot(x,np.tan(x),'g')

plt.subplot(2, 2, 3)
plt.plot(x,np.exp(x))
plt.title("e^x")

plt.subplot(2, 2, 4)
plt.hist(np.tan(x), bins=35)
plt.title("Histogram of tan(x)")

plt.tight_layout()
plt.show()
```
