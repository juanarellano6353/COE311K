```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

cs = CubicSpline(x, y, bc_type='natural')

coeffs = np.polyfit(x, y, 3)
poly = np.poly1d(coeffs)

x_interp = np.linspace(0, 5, 100)

y_spline = cs(x_interp)
y_poly = poly(x_interp)

plt.plot(x, y, 'bo', label="Data points")
plt.plot(x_interp, y_spline, 'r-', label="Cubic Spline Interpolation")
plt.plot(x_interp, y_poly, 'g--', label="Cubic Polyfit")
plt.title("Polyfit vs CubicSpline")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# The cubic spline goes exactly through each data point and smoothly connects them. 
# The polyfit curve is one overall polynomial, so it does not pass through every point exactly.
# Because of this, the spline follows the pattern of the data more closely, 
# while the polynomial fit gives a smoother overall approximation.

#AI USAGE DISCLOSURE:
# I used AI to double check my syntax and help me with the plot.
# The main approach of the code come from the examples shown in the
# lecture slides.
#
# AI Tool Used: ChatGPT
# Prompt Used:
# "Help me check my code which compares polyfit and cubic spline interpolation."
```
