```python
import numpy as np

true_pi = np.pi

terms = 5000
print_every = 100

partial_sum = 0.0 

for k in range(terms):
    term = ((-1)**k) / (2*k + 1)    
    partial_sum += term
    pi_est = 4.0 * partial_sum

    if (k + 1) % print_every == 0:
        rel_error = abs(pi_est - true_pi) / abs(true_pi)
        print(f"Terms: {k+1:4d} | pi_est: {pi_est:.15f} | relative error: {rel_error:.6e}")
        

# I used an AI tool to help me with the syntax in order to code the correct form of the Leibniz Series of pi.
# AI Tool Used: ChatGPT
# AI Prompt Used:
# "Help me with the syntax for the Leibniz Series for pi in Python, run 5,000 terms, print relative error every 100 terms."
#
# The error seems to decrease, although it's not completely smooth after 
# every term with the convergence being kind of slow even with 5,000 terms.
# The error is still somewhat noticeable, at a VERY small degree, especially compared to the Monte Carlo method we had used before.
```
