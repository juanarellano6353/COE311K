```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]

    return np.array(C)


sizes = [3]
for _ in range(7):
    sizes.append(sizes[-1] * 2)

naive_times = []
numpy_times = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    start = time.perf_counter()
    C_naive = naive_matrix_multiplication(A, B)
    end = time.perf_counter()
    naive_times.append(end - start)

    start = time.perf_counter()
    C_numpy = np.dot(A, B)
    end = time.perf_counter()
    numpy_times.append(end - start)

    print(f"Matrix size: {n}x{n}")
    print(f"Naive time: {naive_times[-1]:.6f} seconds")
    print(f"Numpy time: {numpy_times[-1]:.6f} seconds")
    print()

matrix_numbers = np.arange(1, 9)

plt.figure(figsize=(8, 5))
plt.plot(matrix_numbers, naive_times, marker='o', label='Naive (custom)')
plt.plot(matrix_numbers, numpy_times, marker='s', label='Optimized (numpy)')
plt.xlabel("Matrix Number")
plt.ylabel("Runtime (seconds)")
plt.title("Naive vs Optimized Matrix Multiplication Runtime")
plt.xticks(matrix_numbers)
plt.grid(True)
plt.legend()
plt.show()
```
